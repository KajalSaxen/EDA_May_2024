- iterate through directory

- if that values== object types, then keep that corresponding key in one list

- else keep that variable in other listm

```python
data_types=Visa_df.dtypes
dict(data_types)
```

```python
cat_list=[]
num_list=[]
for i,j in dict(data_types).items():
    print(i,j)
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(1)
```

```python
cat_list=[i for i,j in dict(data_types).items() if j=='object']
num_list=[i for i,j in dict(data_types).items() if j!='object']
cat_list,num_list
```

**take**

- first argument:indices
- indices will take some value in array: in list 
- it tries to extract the data 
- that value we provide represents columns or rows??
- Axis=0 represents about rows
- Asis=1 represents about columns 
- by default axis mentioned as zero 

```python
# case-1:
Visa_df.take([100,200,300])
## it retrives 3 rows which having index 100,200,300
```
